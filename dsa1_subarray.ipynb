Kamal is a software developer and he's working on a new feature for an e-commerce website. The website has a list of prices for different products, and Kamal needs to find the maximum price of all products up to a given index.

Problem Constraints
1 <= N <= 105
1 <= A[i] <= 109


Input Format

Return an array of integers where the i-th element of the array represents the maximum value occurring in the subarray from 0 to i in the input array A.

```python
class SolutionPrefixMax:
    def __init__(self,array_p):
        self.A = array_p

    def prefix_max(self):
        n = len(self.A)
        P = []
        P.append(self.A[0])
        for i in range(1,n):
            P.append(max(P[i-1],self.A[i]))
        return P
    
A = [9, 7, 6, 18, 2]
s3 = SolutionPrefixMax(A)
s3.prefix_max()
```

Problem Constraints

1 <= A[i] <= 109


Input Format

Output Format
Return an array of integers denoting the suffix maximum of the given array.


Example Input
Input 1:

A = [12, 5, 6, 7]
Input 2:

A = [15, 9, 7, 11, 10]


Example Output
Output 1:

[12, 7, 7, 7]
Output 2:

[15, 11, 11, 11, 10]


Example Explanation
Explanation 1:

The suffix maximum array of [12, 5, 6, 7]] is [12, 7, 7, 7].
Explanation 2:

The suffix maximum array of [15, 9, 7, 11, 10] is [15, 11, 11, 11, 10].

```python
class SolutionSuffixMax:
    def __init__(self,array_p):
        self.A = array_p

    def suffix_max(self):
        n = len(self.A)
        S=[]
        S.append(self.A[n-1])
        for i in reversed(range(n-1)):
            S.append(max(S[n-2-i],self.A[i]))
        return S[::-1]

A = [12, 5, 6, 7]       
s4 = SolutionSuffixMax(A)
s4.suffix_max()
```
