Subarray introduction
total no. of subarrays = n +(n-1) + (n-2) + ..+3+2+1 = n*(n+1)/2

```python
A = [1,2,3,4,5]
class SolutionSubarrayIntro:
    def __init__(self,array_p):
        self.A = array_p

    def subarray_print(self,start,end):
        ans =[]
        for i in range(start,end+1):
            ans.append(self.A[i])
        return ans
    
    def subarray_all_print(self):
        n = len(self.A)
        ans =[]
        for i in range(n): # start   
            for j in range(i,n): # end 
                row=[]
                for k in range(i,j+1):
                    row.append(A[k])
                ans.append(row)
        return ans
    
    def subarray_all_print_newline(self):
        n = len(self.A)
        for i in range(n):
            for j in range(i,n):
                #row=[]
                for k in range(i,j+1):
                    #row.append(A[k])
                    print(A[k],end=" ")
                print()
    def subarray_sum_bruteforce(self):
        n = len(self.A)
        
        for i in range(n):# start index 
            for j in range(i,n): # end index
                sum_s=0
                for k in range(i,j+1):
                    sum_s+= A[k]
                print(sum_s)

    def subarray_sum_prefixsum(self):
        def prefix_sum():
            n = len(self.A)
            sum_p = 0
            P = []
            for i in range(n):
                sum_p += self.A[i]
                P.append(sum_p)
            return P
        P = prefix_sum()
        n = len(self.A)
        for i in range(n):
            for j in range(i,n):
                if i == 0:
                    print(P[j])
                else:
                    print(P[j]-P[i-1])

    def subarray_sum_carryforward(self,start,end):
        #n = len(self.A)
        sum_c = 0
        for i in range(start,end+1):
            sum_c += self.A[i]
            print(sum_c)

    
    def subarray_all_sum_carryforward(self):
        n = len(self.A)
        for i in range(n):# i = start index 
            self.subarray_sum_carryforward(i,n-1)

    def subarray_all_sum_carryforward1(self):
        n = len(self.A)
        ans =[]
        for i in range(n):
            sum_c = 0
            for j in range(i,n):
                sum_c += self.A[j]
                ans.append(sum_c)
                #print(sum_c)
        return ans
    
    def subarray_maxsum_carryforward(self):
        n = len(self.A)
        ans = float('-inf')
        for i in range(n):
            sum_c = 0
            for j in range(i,n):
                sum_c += self.A[j]
                ans = max(ans,sum_c)

        return ans
    
    def subarray_minsum_carryforward(self):
        n = len(self.A)
        ans = float('+inf')
        for i in range(n):
            sum_c = 0
            for j in range(i,n):
                sum_c += self.A[j]
                ans = min(ans,sum_c)

        return ans
    
    def subarray_evensum_carryforward(self):
        n = len(self.A)
        count = 0
        for i in range(n):
            sum_c = 0
            for j in range(i,n):
                sum_c += self.A[j]
                if (sum_c % 2 == 0):
                    count += 1
                #print(sum_c)
        return count
    
    def subarray_oddsum_carryforward(self):
        n = len(self.A)
        count = 0
        for i in range(n):
            sum_c = 0
            for j in range(i,n):
                sum_c += self.A[j]
                if (sum_c % 2 != 0):
                    count += 1
                #print(sum_c)
        return count
    
    def subarray_evenlengthsum_gt_B_carryforward(self,B):
        n = len(self.A)
        count = 0
        for i in range(n):
            sum_c = 0
            for j in range(i,n):
                sum_c += self.A[j]
                length = j-i + 1
                if (length % 2 == 0 and sum_c > B):
                    count += 1
                #print(sum_c)
        return count







    
s1 = SolutionSubarrayIntro(A)
#s1.subarray_print(2,4)
#s1.subarray_all_print()
#s1.subarray_all_print_newline()
#s1.subarray_sum_bruteforce()
#print("Check")
#s1.subarray_sum_prefixsum()
#s1.subarray_sum_carryforward(2,4)
#s1.subarray_all_sum_carryforward()
#s1.subarray_all_sum_carryforward1()
#s1.subarray_maxsum_carryforward()
#s1.subarray_minsum_carryforward()
s1.subarray_evensum_carryforward()
s1.subarray_oddsum_carryforward()
s1.subarray_evenlengthsum_gt_B_carryforward(3)
```
