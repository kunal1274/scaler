```python
A = [2, 1, 8, 3, 9, 6]
B = [[0, 3],[3, 5],[1, 3],[2, 4]]

class SolutionEvenRangeSum:
    def __init__(self,array_p,range_queries):
        self.A = array_p
        self.B = range_queries

    def prefix_sum_even(self):
        n = len(self.A)
        P =[]
        sum_a = 0
        for i in range(n):
            if A[i]%2 == 0 :
                sum_a+=1
            P.append(sum_a)

        return P
    
    def prefix_sum_odd(self):
        n = len(self.A)
        P =[]
        sum_a = 0
        for i in range(n):
            if A[i]%2 != 0 :
                sum_a+=1
            P.append(sum_a)

        return P
    
    def count_even(self):
        P = self.prefix_sum_even()
        ans =[]
        for left,right in self.B:
            if left == 0 :
                ans.append(P[right])
            else:
                ans.append(P[right] - P[left-1])
        return ans
    
    def count_odd(self):
        P = self.prefix_sum_odd()
        ans =[]
        for left,right in self.B:
            if left == 0 :
                ans.append(P[right])
            else:
                ans.append(P[right] - P[left-1])
        return ans

    
s2 = SolutionEvenRangeSum(A,B)
s2.prefix_sum_even()
s2.count_even()
s2.count_odd()
        
```

Problem Description
You are given an array A of integers of size N.

Your task is to find the equilibrium index of the given array

The equilibrium index of an array is an index such that the sum of elements at lower indexes is equal to the sum of elements at higher indexes.

If there are no elements that are at lower indexes or at higher indexes, then the corresponding sum of elements is considered as 0.

Note:

Array indexing starts from 0.
If there is no equilibrium index then return -1.
If there are more than one equilibrium indexes then return the minimum index.


Problem Constraints
1 <= N <= 105
-105 <= A[i] <= 105


Input Format
First arugment is an array A .


Output Format
Return the equilibrium index of the given array. If no such index is found then return -1.


Example Input
Input 1:
A = [-7, 1, 5, 2, -4, 3, 0]
Input 2:

A = [1, 2, 3]


Example Output
Output 1:
3
Output 2:

-1


Example Explanation
Explanation 1:
i   Sum of elements at lower indexes    Sum of elements at higher indexes
0                   0                                   7
1                  -7                                   6
2                  -6                                   1
3                  -1                                  -1
4                   1                                   3
5                  -3                                   0
6                   0                                   0

3 is an equilibrium index, because: 
A[0] + A[1] + A[2] = A[4] + A[5] + A[6]
Explanation 1:

i   Sum of elements at lower indexes    Sum of elements at higher indexes
0                   0                                   5
1                   1                                   3
2                   3                                   0
Thus, there is no such index.

```python
class SolutionEquilibriumIndex:
    """
    Objective : Finding the Equilibrium index for which left side is equal to the right side of the index.
    Logic :
    1. First find the prefix sum .
    2. starting with 1st element of the prefix sum . when index is zero then left side = 0 and right side is P[n-1]-P[index]. if both equal then add to the output.
    if index not equal to zero then left side is P[index - 1] and right side is P[n-1] - P[index]
    if index equal to n-1 or the last index of the Prefix sum array then ls = P[index -1 ] and right side = 0
    """
    def __init__(self,array_p):
        self.A = array_p
        

    def prefix_sum(self):
        n = len(self.A)
        P =[]
        P.append(self.A[0])
        for i in range(1,n):
            P.append(P[i-1] + self.A[i])

        return P
    
    def prefix_sum1(self):
        sum_a =0 
        P =[]
        n = len(self.A)
        for i in range(n):
            sum_a += self.A[i]
            P.append(sum_a)
        return P
    
    def equilibrium_index(self):
        P = self.prefix_sum()
        n = len(P)
        ans=[]
        for i in range(n):
            if i == 0:
                left_side = 0
                right_side = P[n-1]-P[i]
                if left_side == right_side:
                    ans.append(i)
            elif i == n-1:
                left_side = P[i-1]
                right_side = 0
                if left_side == right_side :
                    ans.append(i)
            else:
                left_side = P[i-1]
                right_side = P[n-1] - P[i]
                if left_side == right_side :
                    ans.append(i)

        return ans
    
    def is_equilibrium(self):
        ans = self.equilibrium_index()
        if len(ans) > 0:
            return min(ans)
        else:
            return -1

A = [-7, 1, 5, 2, -4, 3, 0]
s3 = SolutionEquilibriumIndex(A)
#s3.prefix_sum()
s3.is_equilibrium()
A1 = [1,2,3]
s3_1 = SolutionEquilibriumIndex(A1)
s3_1.is_equilibrium()
```

Note : Use 0-based indexing


Problem Constraints
1 <= N <= 105
1 <= Q <= 105
1 <= A[i] <= 100
0 <= B[i][0] <= B[i][1] < N


Input Format
First argument A is an array of integers.
Second argument B is a 2D array of integers.


Output Format
Return an array of integers.


Example Input
Input 1:
A = [1, 2, 3, 4, 5]
B = [   [0,2] 
        [1,4]   ]
Input 2:
A = [2, 1, 8, 3, 9]
B = [   [0,3] 
        [2,4]   ]


Example Output
Output 1:
[4, 8]
Output 2:
[10, 17]


Example Explanation
For Input 1:
The subarray for the first query is [1, 2, 3] whose sum of even indices is 4.
The subarray for the second query is [2, 3, 4, 5] whose sum of even indices is 8.
For Input 2:
The subarray for the first query is [2, 1, 8, 3] whose sum of even indices is 10.
The subarray for the second query is [8, 3, 9] whose sum of even indices is 17.

```python
class SolutionSumIndices:
    def __init__(self,array_p,range_queries):
        self.A = array_p
        self.B = range_queries

    def prefix_sum_even(self):
        n = len(self.A)
        P =[]
        sum_a = 0
        for i in range(n):
            if A[i]%2 == 0 :
                sum_a+=1
            P.append(sum_a)

        return P
    
    def prefix_sum_odd(self):
        n = len(self.A)
        P =[]
        sum_a = 0
        for i in range(n):
            if A[i]%2 != 0 :
                sum_a+=1
            P.append(sum_a)

        return P
    
    def count_even(self):
        P = self.prefix_sum_even()
        ans =[]
        for left,right in self.B:
            if left == 0 :
                ans.append(P[right])
            else:
                ans.append(P[right] - P[left-1])
        return ans
    
    def count_odd(self):
        P = self.prefix_sum_odd()
        ans =[]
        for left,right in self.B:
            if left == 0 :
                ans.append(P[right])
            else:
                ans.append(P[right] - P[left-1])
        return ans
    
    def prefix_sum_indices(self):
        n = len(self.A)
        PE=[]
        PO =[]
        sum_e = 0
        sum_o = 0
        for i in range(n):
            if i%2 == 0 :# even indices
                sum_e+=self.A[i]
                PE.append(sum_e)
                PO.append(sum_o)
            else: # odd indices
                sum_o+=self.A[i]
                PO.append(sum_o)
                PE.append(sum_e)
        return (PE,PO)
    
    def sum_even_indices(self):
        P = self.prefix_sum_indices()[0]
        n = len(P)
        ans =[]
        for left,right in self.B:
            if left == 0:
                ans.append(P[right])
            else:
                ans.append(P[right] - P[left-1])
        return ans
    
    def sum_odd_indices(self):
        P = self.prefix_sum_indices()[1]
        n = len(P)
        ans =[]
        for left,right in self.B:
            if left ==0:
                ans.append(P[right])
            else:
                ans.append(P[right] - P[left-1])
        return ans
        


    
A = [2, 1, 8, 3, 9]
B = [   [0,3],
        [2,4]   ]
s4 = SolutionSumIndices(A,B)
s4.prefix_sum_indices()
s4.sum_even_indices()
s4.sum_odd_indices()
```

Given an array of integers A, find and return the product array of the same size where the ith element of the product array will be equal to the product of all the elements divided by the ith element of the array.

Note: It is always possible to form the product array with integer (32 bit) values. Solve it without using the division operator.


Input Format

The only argument given is the integer array A.
Output Format

Return the product array.
Constraints

2 <= length of the array <= 1000
1 <= A[i] <= 10
For Example

Input 1:
    A = [1, 2, 3, 4, 5]
Output 1:
    [120, 60, 40, 30, 24]

Input 2:
    A = [5, 1, 10, 1]
Output 2:
    [10, 50, 5, 50]

```python
class SolutionProductPuzzle:
    def __init__(self,array_p):
        self.A = array_p

    def prefix_product(self):
        P=[]
        prod = 1
        n = len(self.A)
        for i in range(n):
            prod *= self.A[i]
            P.append(prod)
        return P
    
    def product_puzzle(self):
        P = self.prefix_product()
        n = len(P)
        ans =[]
        for i in range(n):
            if i ==0:
                ans.append(P[n-1]//P[0])
            else:
                ans.append((P[n-1]*P[i-1])//P[i])
        return ans

A = [1, 2, 3, 4, 5]   
s5 = SolutionProductPuzzle(A)
s5.prefix_product()
s5.product_puzzle()
```
